```python
import xml.etree.ElementTree as ET
inheritence_patterns = ["X-linked dominant", "Autosomal dominant", "Y-linked"]
# Parse the XML file
tree = ET.parse('data/source_files/Natural History Ages.xml')
root = tree.getroot()

# List to store the disease names
autosomal_dominant_diseases = []
x_linked_dominant_diseases = []
y_linked_diseases = []
# Loop through all disorder elements
for disorder in root.findall('.//Disorder'):
    # Find the inheritance list
    inheritance_list = disorder.find('.//TypeOfInheritanceList')
    orpha_code = disorder.find('.//OrphaCode')
    if inheritance_list is not None:
        for inheritance in inheritance_list.findall('.//Name'):
            if inheritance.text in inheritence_patterns:
                # Find the disease name
                name = disorder.find('.//Name')
                if name is not None:
                    if (inheritance.text == "Autosomal dominant"):
                        autosomal_dominant_diseases.append(name.text)
                    elif (inheritance.text == "X-linked dominant"):
                        x_linked_dominant_diseases.append(name.text)
                    elif (inheritance.text == "Y-linked"):
                        y_linked_diseases.append(name.text)

# Print the list of autosomal dominant diseases
print(autosomal_dominant_diseases)
print(len(autosomal_dominant_diseases))
print(x_linked_dominant_diseases)
print(len(x_linked_dominant_diseases))
print(y_linked_diseases)
print(len(y_linked_diseases))
```
